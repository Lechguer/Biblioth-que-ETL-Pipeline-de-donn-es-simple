# 📊 Analyse de la Bibliothèque Personnelle
> Ce notebook analyse les données produites par le pipeline ETL.

```python
import pandas as pd

# Charger le fichier final produit par le pipeline
df = pd.read_csv("output/rapport_final.csv")

print(f"Nombre de livres : {df.shape[0]}")
df.head()
```

## 1. Statistiques générales

```python
# Résumé statistique des colonnes numériques
df[['annee', 'note', 'age_livre']].describe().round(2)
```

## 2. Répartition par genre

```python
# Nombre de livres par genre
nb_par_genre = df['genre'].value_counts()
print(nb_par_genre)

# Visualisation rapide
nb_par_genre.plot(kind='bar', title='Nombre de livres par genre', color='steelblue')
```

## 3. Note moyenne par genre

```python
# Moyenne des notes par genre, triée du meilleur au moins bon
note_par_genre = df.groupby('genre')['note'].mean().sort_values(ascending=False).round(2)
print(note_par_genre)

note_par_genre.plot(kind='barh', title='Note moyenne par genre', color='coral')
```

## 4. Top 5 des meilleurs livres

```python
# Les 5 livres les mieux notés
top5 = df[['titre', 'auteur', 'note', 'genre']].sort_values('note', ascending=False).head(5)
print(top5.to_string(index=False))
```

## 5. Auteurs les plus représentés

```python
# Auteurs avec le plus de livres dans la bibliothèque
df['auteur'].value_counts().head(10)
```
